## Run

```python
def worker(row):
    try:
        result = pmc_tables.download_archive(
            op.join(SOURCE_DIR, row['archive_file_path']),
            op.join(OUTPUT_DIR, row['output_file_path']),
            pmc_type=row['pmc_type'],
            pmc_id=row['pmc_id'],
            source_url=row['source_url']
        )
    except Exception as e:
        result = {'error': str(e)}
    return result
```

```python
N_PROCS = 16

with concurrent.futures.ProcessPoolExecutor(N_PROCS) as pool:
    rows = [
        row._asdict()
        for row in to_download.itertuples()
        if op.join(SOURCE_DIR, row.archive_file_path) in source_files
        and not op.isfile(op.join(OUTPUT_DIR, row.output_file_path))
    ]
    futures = list(pool.map(worker, rows, chunksize=1_000))
```
